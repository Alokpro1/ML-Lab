```python
import math
import matplotlib.pyplot as plt
import numpy as np
from scipy.special import factorial
```

```python
x=np.arange(100)
```

```python
def normal(sigma,mean):
    data_y= (1.0 / (sigma * math.sqrt(2*math.pi)))*np.exp(-0.5*((x - mean) / sigma) ** 2)
    return data_y
```

```python
def binomial(n,p):
    fact_x=factorial(x, exact=False)
    fact_nx=factorial(n-x, exact=False)
    data_y=(factorial(n, exact=False)/(fact_x*fact_nx))*(p**x)*((1-p)**(n-x))
    return data_y
    
```

```python
def gamma(alpha,beta):
    return ((beta**alpha)*(x**(alpha-1))*np.exp(-beta*x))/(factorial(x-1, exact=False))
```

```python
def expo(lamb):
    return lamb*np.exp(-lamb*x)
```

```python
data_yn=normal(15,50)

plt.plot(x, data_yn, label='normal')

plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Normal")
plt.legend()
```

```python

data_yg=gamma(6,0.2)

plt.plot(x, data_yg, label='gamma')

plt.xlabel('x label')
plt.ylabel('y label')
plt.title("gamma")
plt.legend()
```

```python

data_ye=expo(0.9)

plt.plot(x, data_ye, label='expo')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("expo")
plt.legend()
```

```python
data_yb=binomial(40,0.4)

plt.plot(x, data_yb, label='binomial')

plt.xlabel('x label')
plt.ylabel('y label')
plt.title("binomial")
plt.legend()
```
